```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator, load_img
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_yes = load_img('brain_tumor_dataset/yes/Y11.jpg')
fig = plt.figure(figsize=(16,8))
tumor = fig.add_subplot(1,3,1)
plt.imshow(img_yes)
tumor.set_xlabel('tumor')

img_no = load_img('brain_tumor_dataset/no/N3.jpg')
normal = fig.add_subplot(1,3,2)
plt.imshow(img_no)
normal.set_xlabel("Normal")
plt.show()
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python
test_datagen = ImageDataGenerator(rescale=1. / 255)
```

```python
training_set = train_datagen.flow_from_directory('brain_tumor_dataset',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

```python
test_set = test_datagen.flow_from_directory('brain_tumor_dataset',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 30)
```
